```python
import scipy.io
import numpy as np
import os, re
import cv2
import csv
import pickle
```

```python
def atoi(text):
    return int(text) if text.isdigit() else text
def natural_keys(text):
    return [ atoi(c) for c in re.split('(\d+)', text) ]
```

```python
dataPath = '/media/yonsei/4TB_HDD/dataset/suncg_data/'
metaDataPath = os.path.join(dataPath, 'metadata')
modelCategoryMappingPath = os.path.join(metaDataPath, 'ModelCategoryMapping_modified.csv')
modelNetObjDataPath = os.path.join(dataPath, 'modelNet')
objClassList = os.listdir(modelNetObjDataPath)
with open(modelCategoryMappingPath, 'r') as dest_f:
    data_iter = csv.reader(dest_f,delimiter=",",quotechar='"')
    metadata = [metadata for metadata in data_iter]
metadata = np.asarray(metadata)
objClassDict = dict()
for objClass in objClassList:
    if objClass not in objClassDict:
        objClassDict[objClass] = []
for info in range(len(metadata)):
    if metadata[info, 11] not in objClassDict and metadata[info,11] != 'final_class':
        objClassDict[metadata[info,11]] = []
        
for objClass in objClassList:
    for trainOrTest in ['train', 'test']:
        objPerClassPath = os.path.join(modelNetObjDataPath, objClass, trainOrTest)
        objPerClassList = os.listdir(objPerClassPath)
        for obj in objPerClassList:
            objClassDict[objClass] += [os.path.join('modelNet', objClass, trainOrTest, obj, obj)]
relatedClassList = np.column_stack((metadata[:,1], metadata[:,10], metadata[:,11]))
for info in range(len(metadata)):
    if metadata[info, 11] in objClassDict:
        suncgObjPath = os.path.join("object",metadata[info,1],metadata[info,1])
        objClassDict[metadata[info,11]].append(suncgObjPath)
```

```python
for key in objClassDict.keys():
    print key, len(objClassDict[key])
```

```python
objClassAndInst = dict()
classIdx = 0
for key in objClassDict.keys():
    objListPerClass = objClassDict[key]
    instIdx = 0
    for obj in objListPerClass:
        objClassAndInst[obj] = [classIdx, instIdx]
        instIdx+=1
    classIdx+=1
with open(os.path.join(metaDataPath, 'objClassAndInst.pkl'), 'wb') as objClassAndInstFile:
    pickle.dump(objClassAndInst, objClassAndInstFile)
    objClassAndInstFile.close()
```

```python
# read python dict back from the file
with open(os.path.join(metaDataPath,'objClassAndInst.pkl'), 'rb') as objClassAndInstFile:
    mydict = pickle.load(objClassAndInstFile)
    objClassAndInstFile.close()
for key in mydict.keys():
    print key, mydict[key]
```
